```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

import tensorflow as tf
from tensorflow.keras import Sequential # It is build to build ann. Anything which is capital is a class
from tensorflow.keras.layers import Dense # it is used to add hidden layers
from sklearn.metrics import classification_report #evaluation
```

```python
df=pd.read_csv('/content/diabetes.csv')
```

```python
df
```

```python
df.info()
```

```python
df.drop(['Pregnancies'],axis=1,inplace=True)
```

```python
df
```

```python
df.describe()
```

```python
from scipy.stats import skew
```

```python
df['Insulin']=skew(df['Insulin'])
```

```python
df.describe()
```

```python
x=df.iloc[:,:-1]
y=df['Outcome']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20,random_state=1)
```

```python
xtrain
```

```python
xtest
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.fit_transform(xtest)
```

```python
# init model
ann=Sequential()

#step2: add layers into model
ann.add(Dense(units=12,activation='relu')) # created one hidden layer
ann.add(Dense(units=1,activation='sigmoid')) # output layer

# step3 establish connection b/w layers
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

#step4 train the model
ann.fit(xtrain,ytrain,batch_size=35,epochs=100)
```

```python

ypred=ann.predict(xtest)
```

```python
ypred=ypred>0.5
ypred=np.where(ypred<0.5,0,1)
ypred
```

```python
print(classification_report(ytest,ypred))
```
